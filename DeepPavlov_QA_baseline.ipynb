## Russian Open-domain QA with DeepPavlov Starter Code

This file contains starter code for the open-domain QA in Russian based on Wikipedia. The system is developed by DeepPavlov project. More on the system: http://docs.deeppavlov.ai/en/master/features/skills/odqa.html

Due to version differences in dependencies, we have added code to uninstall  tensorflow and install tensorflow-gpu==1.14.0, which makes DeepPavlov library currently (as of 6 February 2022) working in Google Colab. Importing 'export_saved_model' is a test that the error due to which DeepPavlov could not be installed is gone. If you are working in Google Colab, first, install tensorflow-gpu==1.14.0. Then restart the runtime, comment out  and run the cell with importing 'export_saved_model' and teh rest of the cells.

Two tips on how this model can be improved:
1. Preprocessing questions so that they resemble typical WH-questions.
2. Training DeepPavlov with the Russian QA Jeopardy! dataset.
